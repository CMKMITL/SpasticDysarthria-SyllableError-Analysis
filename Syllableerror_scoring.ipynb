# **Step 1 : Transpose four Sentences Into Dataframes**

```python
import pandas as pd
from scipy import stats

###Import CSV file of separated syllables for all sentences###
from google.colab import files

file_separated_syllables = files.upload()
csv_separated_syllables = next(iter(file_separated_syllables))

df = pd.read_csv(csv_separated_syllables)

print(df)
```

```python
###Build dataframe showing separated syllables for each sentence###
sen_1 = df.columns[2:11]
sen_2 = df.columns[11:20]
sen_3 = df.columns[20:29]
sen_4 = df.columns[29:45]

df.loc[:,sen_1]
df.loc[:,sen_2]
df.loc[:,sen_3]
df.loc[:,sen_4]
```

# **Step 2 : Evaluate Syllable Error Scores by Sentence (defined criteria is shown)**

```python
###Count Syllable_errors for sentence 1###

sen_1
s = [x for x in sen_1 if 's' in x]
e = [x for x in sen_1 if 'e' in x]

t1_1 = ['ปู']
t1_2 = ['ม้า']
t1_3 = ['วิ่ง']
t1_4 = ['ไป']
t1_5 = ['มา']
t1_6 = ['บน']
t1_7 = ['ใบ']
t1_8 = ['ไม้']

score_s1 = [];
total = [t1_1,t1_2,t1_3,t1_4,t1_5,t1_6,t1_7,t1_8]

s = [x for x in sen_1 if 's' in x]
e = [x for x in sen_1 if 'e' in x]

this_df = df.loc[:,sen_1];
for i in range(df.shape[0]):
    x = 0;
    for j in range(len(s)):
        if this_df.loc[i,s[j]] in total[j]:
            x += 0
        else:
            x += 1

    for j in range(len(e)):
        if str(this_df.loc[i,e[j]]) != '0':
            x += 1
        else:
            x += 0

    score_s1.append(x)
```

```python
###Count Syllable_errors for sentence 2###

t2_1 = ['ชาว']
t2_2 = ['ไร่','ไล่']
t2_3 = ['ตัด']
t2_4 = ['ต้น']
t2_5 = ['สน']
t2_6 = ['ทำ','ธรรม']
t2_7 = ['ท่อน']
t2_8 = ['ซุง']

score_s2 = [];
total = [t2_1,t2_2,t2_3,t2_4,t2_5,t2_6,t2_7,t2_8]

s = [x for x in sen_2 if 's' in x]
e = [x for x in sen_2 if 'e' in x]

this_df = df.loc[:,sen_2];
for i in range(df.shape[0]):
    x = 0;
    for j in range(len(s)):
        if this_df.loc[i,s[j]] in total[j]:
            x += 0
        else:
            x += 1

    for j in range(len(e)):
        if str(this_df.loc[i,e[j]]) != '0':
            x += 1
        else:
            x += 0
    score_s2.append(x)
```

```python
###Count Syllable_errors for sentence 3###

t3_1 = ['อี','อีก']
t3_2 = ['กา']
t3_3 = ['คอย','คอยน์']
t3_4 = ['คาบ']
t3_5 = ['งู']
t3_6 = ['คาบ']
t3_7 = ['ไก่']

score_s3 = [];
total = [t3_1,t3_2,t3_3,t3_4,t3_5,t3_6,t3_7]

s = [x for x in sen_3 if 's' in x]
e = [x for x in sen_3 if 'e' in x]

this_df = df.loc[:,sen_3];
for i in range(df.shape[0]):
    x = 0;
    for j in range(len(s)):
        if this_df.loc[i,s[j]] in total[j]:
            x += 0
        else:
            x += 1

    for j in range(len(e)):
        if str(this_df.loc[i,e[j]]) != '0':
            x += 1
        else:
            x += 0

    score_s3.append(x)
```

```python
###Count Syllable_errors for sentence 4###

t4_1 = ['เพียง']
t4_2 = ['แค่']
t4_3 = ['ฝน']
t4_4 = ['ตก']
t4_5 = ['ลง']
t4_6 = ['ที่']
t4_7 = ['หน้า']
t4_8 = ['ต่าง']
t4_9 = ['ใน']
t4_10 = ['บาง']
t4_11 = ['ครา','คา']


score_s4 = [];
total = [t4_1,t4_2,t4_3,t4_4,t4_5,t4_6,t4_7,t4_8,t4_9,t4_10,t4_11]

s = [x for x in sen_4 if 's' in x]
e = [x for x in sen_4 if 'e' in x]
this_df = df.loc[:,sen_4];
for i in range(df.shape[0]):
    x = 0;
    for j in range(len(s)):
        if this_df.loc[i,s[j]] in total[j]:
            x += 0
        else:
            x += 1

    for j in range(len(e)):
        if str(this_df.loc[i,e[j]]) != '0':
            x += 1
        else:
            x += 0

    score_s4.append(x)
```

```python
###Build dataframe showing syllable error scores for each sentence in 74 subjects###
df_score = pd.DataFrame(list(zip(score_s1,score_s2,score_s3,score_s4)), columns = ['error_1','error_2','error_3','error_4'])
print(df_score)

df_result = pd.concat([df.iloc[:,0:2],df_score],axis=1)
df_result
```

```python
###Save the dataframe###
df_result.to_csv('Syllableerror_score.csv', index=False)

from google.colab import files
files.download('Syllableerror_score.csv')
```
